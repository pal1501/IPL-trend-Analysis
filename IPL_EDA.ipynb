DATA ANALYSIS

```python
# Correlation Analysis
numerical_features = ['result_margin', 'id', 'year']
correlation_matrix = regular_df[numerical_features].corr()
plt.figure(figsize=(12, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix of Numerical Features')
plt.show()

# Feature Relationships with Position
plt.figure(figsize=(15, 8))
for i, feature in enumerate(numerical_features):
  plt.subplot(3, 3, i + 1)
  sns.boxplot(x='winner', y=feature, data=regular_df)
  plt.title(f'{feature} vs winner')
plt.tight_layout()
plt.show()

for winner in regular_df['winner'].unique():
    print(f"\nDescriptive Statistics for winner {winner}:")
    display(regular_df[regular_df['winner'] == winner][numerical_features].describe())

# Team Performance Over Time
plt.figure(figsize=(10, 6))
for feature in ['result_margin', 'year']:
    df_yearly_avg = regular_df.groupby('year')[feature].mean()
    plt.plot(df_yearly_avg.index, df_yearly_avg.values, label=feature)
plt.xlabel('year')
plt.ylabel('Average Result Margin/Year')
plt.title('Average Result Margin and Year Over Time')
plt.legend()
plt.show()
```

DATA VISUALIZATION

```python
# Define numerical features
numerical_features = ['Points', 'Net Run Rate', 'Matches Played', 'Won', 'Lost']

# -----------------------
# Histograms
# -----------------------
plt.figure(figsize=(16, 10))
for i, col in enumerate(numerical_features):
    plt.subplot(2, 3, i + 1)
    sns.histplot(playoffs_df[col], kde=True, bins=10, color='skyblue')
    plt.title(f'Distribution of {col}')
plt.tight_layout()
plt.show()

# -----------------------
# Box plots vs Position
# -----------------------
plt.figure(figsize=(16, 10))
for i, col in enumerate(numerical_features):
    plt.subplot(2, 3, i + 1)
    sns.boxplot(x='Position', y=col, data=playoffs_df, palette='pastel')
    plt.title(f'{col} vs Position')
plt.tight_layout()
plt.show()

# -----------------------
# Scatter plots
# -----------------------
plt.figure(figsize=(16, 5))

plt.subplot(1, 3, 1)
sns.scatterplot(x='Points', y='Net Run Rate', hue='Position', data=playoffs_df)
plt.title('Points vs Net Run Rate')

plt.subplot(1, 3, 2)
sns.scatterplot(x='Won', y='Lost', hue='Position', data=playoffs_df)
plt.title('Won vs Lost')

plt.subplot(1, 3, 3)
sns.scatterplot(x='For Runs', y='Against Runs', hue='Position', data=playoffs_df)
plt.title('For Runs vs Against Runs')

plt.tight_layout()
plt.show()

# -----------------------
# Line plots over Years
# -----------------------
plt.figure(figsize=(10, 5))
for feature in ['Points', 'Net Run Rate']:
    df_yearly_avg = playoffs_df.groupby('Year')[feature].mean()
    plt.plot(df_yearly_avg.index, df_yearly_avg.values, label=feature, marker='o')
plt.xlabel('Year')
plt.ylabel('Average')
plt.title('Average Points and Net Run Rate Over Time')
plt.legend()
plt.grid(True)
plt.show()
```
